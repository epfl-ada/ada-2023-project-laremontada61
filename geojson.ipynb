```python
import pandas as pd
import geopandas as gpd
import json
```

```python
with open("docs/assets/json/geojason.json") as file:
     geo = json.loads(file.read())
```

```python
for feature in geo['features']:
    if feature['properties']['ISO_A2'] == '-99':
        print(feature['properties']['NAME'])
```

```python
cyprus_a2 = "CY"
france = "FR"
norway = "NO"
somaliland = "SO"

for feature in geo['features']:
    if feature['properties']['ISO_A2'] == '-99':
        if feature['properties']['NAME'] == 'Cyprus':
            feature['properties']['ISO_A2'] = cyprus_a2
        elif feature['properties']['NAME'] == 'France':
            feature['properties']['ISO_A2'] = france
        elif feature['properties']['NAME'] == 'Norway':
            feature['properties']['ISO_A2'] = norway
        elif feature['properties']['NAME'] == 'Somaliland':
            feature['properties']['ISO_A2'] = somaliland
```

```python
scores_csv = pd.read_csv("docs/assets/csv/new_ranking_weighted.csv")
with open("docs/assets/json/ranking_weighted.json") as file:
     scores_json = json.loads(file.read())
```

```python
scores_json
```

```python
for brewery_location in scores_csv.brewery_location.unique():
    for feature in geo['features']:
        if brewery_location in feature['properties']['NAME']:
            brewery_data = scores_csv.loc[scores_csv['brewery_location'] == brewery_location]
            feature['properties']['BEST_BREWERY'] = brewery_data['brewery_name'].values[0]
            feature['properties']['GRADE'] = brewery_data['final_grade'].values[0]
            feature['properties']['COLOR'] = brewery_data['colormap'].values[0]

for feature in geo['features']:
    if 'BEST_BREWERY' in feature['properties']:
        print(feature['properties']['NAME'])
        print(feature['properties']['BEST_BREWERY'])
        print("-----------------")
```

```python
with open("docs/assets/json/geo_with_breweries_new.json", "w") as file:
    json.dump(geo, file)
```

```python
cmt = 0
grade_list = []
for feature in geo['features']:
    try: 
        tmp_grade = feature['properties']['GRADE']
        cmt += 1
        grade_list.append(tmp_grade)
    except:
        pass
```

```python
len(grade_list)
```

```python
copy = scores_csv.copy()

for idx, grade in enumerate(grade_list):
    if grade in copy['final_grade'].values:
        copy.drop(copy[copy['final_grade'] == grade].index, inplace=True)
```
