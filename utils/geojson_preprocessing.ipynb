# Pre-processing of GeoJson files

```python
import json
```

# Loading the file

```python
with open("docs/assets/json/ne_110m_admin_0_countries.geojson") as file:
     geo = json.loads(file.read())
```

# Properties

## Showing the properties

```python
print(json.dumps(geo['features'][0], indent=4))
```

## Correct ISO_A2 property

```python
for feature in geo['features']:
    if feature['properties']['ISO_A2'] == '-99':
        print(feature['properties']['NAME'])
```

```python
cyprus_a2 = "CY"
france_a2 = "FR"
norway_a2 = "NO"
somaliland_a2 = "SO"

for feature in geo['features']:
    if feature['properties']['ISO_A2'] == '-99':
        if feature['properties']['NAME'] == 'N. Cyprus':
            feature['properties']['ISO_A2'] = cyprus_a2
        elif feature['properties']['NAME'] == 'France':
            feature['properties']['ISO_A2'] = france_a2
        elif feature['properties']['NAME'] == 'Norway':
            feature['properties']['ISO_A2'] = norway_a2
        elif feature['properties']['NAME'] == 'Somaliland':
            feature['properties']['ISO_A2'] = somaliland_a2
```

## Dropping useless properties for better website performances

```python
properties_to_keep = ['NAME', 'ISO_A2']

for feature in geo['features']:
    properties = feature['properties']
    properties_to_drop = [prop for prop in properties if prop not in properties_to_keep]
    for prop in properties_to_drop:
        properties.pop(prop, None)
```

# Exporting our new GeoJson template

```python
with open("docs/assets/json/geojson_template.json", "w") as file:
    json.dump(geo, file)
```
